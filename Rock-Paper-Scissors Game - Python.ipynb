```python
## Homework
## 1. Pao Ying Chub

import random

# Set up
def play_pyc():
  hands =  ["Rock", "Scissor", "Paper"]
  user_score = 0
  comp_score = 0
  while True :
    user_hand = input("Choose your hand; Rock, Papaer, Scissor or Exit : ")
    comp_hand = random.choice(hands)

    # Check exit
    if user_hand == "Exit":
            print("---Summary Score for both sides---")
            print(f"Total user score is {user_score}")
            print(f"Total computer score is {comp_score}")
            print("Thank you for playing. Goodbye for now!")
            break

    # Check invalid
    if user_hand not in hands:
      print("Invalid input! Please choose 'Rock', 'Paper', or 'Scissor'.")
      continue

    # Display result from choices
    print(f"You chose: {user_hand} | Computer chose: {comp_hand}")

    # Start PYC mechanic
    if user_hand == comp_hand:
        print("Draw !")
    elif (user_hand == "Rock" and comp_hand == "Scissor") or \
         (user_hand == "Paper" and comp_hand == "Rock") or \
         (user_hand == "Scissor" and comp_hand == "Paper"):
        print("You won!")
        user_score = user_score + 1

    else:
        print("You lose")
        comp_score = comp_score + 1
```

```python
play_pyc()
```
