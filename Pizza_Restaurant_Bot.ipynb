```python
## Homework
## 2. Pizza Restaurant Bot

def order_pizza():

  ## bot : greeting
  print("Welcome to Sexy Pizza Restaurant !")
  u_name =  input("What's your name ? ")
  print(f"Welcome to our online restaurant {u_name} ")
  print("There are 3 menus for the online order: Pepperoni Pizza, Fries, Cola ")

  ## bot: there are 3 menus
  menus = ["Pepperoni Pizza", "Fries", "Cola"]
  price = [20, 15, 5]
  pizza_order = 0
  fries_order = 0
  cola_order = 0
  total_order_list = 0
  price_total = 0

  ## user: order menu
  while True:
    item = input("What would you like to order ? Or put 'Done' if done :")

    if (item == "Done"):
      ## bot summarize total order + cost
      total_order_list = pizza_order + fries_order + cola_order
      print("---Summary order list and price---")
      print(f"There are {total_order_list} items that you have ordered." )
      print(f"Pepperoni Pizza {pizza_order} order.")
      print(f"Fries {fries_order} order.")
      print(f"Cola {cola_order} order.")
      print(f"Total price is {price_total} bucks.")
      print("Thank you for coming. Have a good day !")
      break


    # Check invalid ก่อน จากนั้นเป้น Order Mechanic
    if item not in menus:
      print("We don't have that menu ! Please choose 'Pepperoni Pizza', 'Fries', or 'Cola'.")

    elif (item == "Pepperoni Pizza"):
      print("You just ordered 1 Pepperoni Pizza")
      pizza_order = pizza_order + 1
      price_total = price_total + 20

    elif (item == "Fries"):
      print("You just ordered 1 Fries")
      fries_order = fries_order + 1
      price_total = price_total + 15

    elif (item == "Cola"):
      print("You just ordered 1 Cola")
      cola_order = cola_order + 1
      price_total = price_total + 5
```

```python
order_pizza()
```
